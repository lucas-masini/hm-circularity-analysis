Jointure des 2 DataFrames

```python
csv_files = [

    '/Copie de df_2019_clean.csv',
    '/Copie de cust_clean.csv',
]

# Supprimer toutes les tables existantes avant de recréer
tables = con.execute("SHOW TABLES").fetchall()
for t in tables:
    con.execute(f"DROP TABLE IF EXISTS {t[0]}")

# Charger les CSV
for i, file in enumerate(csv_files, start=1):
    table_name = f'table{i}'

    # Forcer la colonne "date" en VARCHAR pour éviter l'erreur
    con.execute(f"""
        CREATE TABLE {table_name} AS
        SELECT * FROM read_csv_auto('{file}', types={{'date': 'VARCHAR'}}, ignore_errors=True)
    """)
    print(f"{table_name} chargé")
```

Anonymisation des données clients avec Faker

```python
from faker import Faker

fake = Faker("fr_FR")

# mapping customer_id -> nom fictif
customer_name_map = {
    cid: fake.name()
    for cid in custo2019["customer_id"].unique()
}

# mapping postal_code -> postal fictif
postal_code_map = {
    pc: fake.postcode()
    for pc in custo2019["postal_code"].unique()
}

# application
custo2019["name_fake"] = custo2019["customer_id"].map(customer_name_map)
custo2019["postal_code_fake"] = custo2019["postal_code"].map(postal_code_map)
```

```python
custo2019clean = custo2019.drop(columns=["postal_code"])
```

```python
# Convertir la colonne 'date' en datetime
custo2019clean['date'] = pd.to_datetime(custo2019['date'], errors='coerce')
```

Mise à l’échelle des prix et des revenus et vérification des statistiques.

```python
# Choix d'un facteur pour ramener les prix à l'échelle réaliste
facteur = 1000

# Transformation des colonnes
custo2019clean['price_real'] = custo2019['price'] * facteur
custo2019clean['revenue_real'] = custo2019['revenue'] * facteur

# Vérification des statistiques
custo2019clean[['price_real', 'revenue_real']].describe()
```

Estimation du genre des clients à partir de leur prénom.

```python
import gender_guesser.detector as gender

d = gender.Detector()
custo2019clean['prenom'] = custo2019clean['name_fake'].str.split().str[0]

# Détecter le genre
custo2019clean['genre'] = custo2019clean['prenom'].apply(lambda x: d.get_gender(x))

# Simplifier
def simplifier(val):
    if val in ['male', 'mostly_male']:
        return 'H'
    elif val in ['female', 'mostly_female']:
        return 'F'
    else:
        return 'U'

custo2019clean['genre'] = custo2019clean['genre'].apply(simplifier)
```
